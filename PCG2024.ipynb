Aqui está sendo feito a mesma coisa com o segundo dataset


```python
df2 = df2.T
df2 = df2.reset_index()
```

```python
df2.columns = ['Ano', 'Área_plantada']
df2 = df2.drop(0)
display(df2)
```

Aqui está sendo selecionado um ano por vez dentro do "for" e calculando a média desse determinado ano e a armazenando dentro de uma lista.

```python
i = 0
year = 13
media = []
for i in range (10):
   ano = df2.loc[df2["Ano"].str.contains(f'/{year}')]
   display(ano)
   media.append((ano.Área_plantada.mean())/10**7)
   year = year + 1

display(media)
```

Aqui está sendo criado o dataframe que possui a média anual da área plantada e a produção de CO2 do estado de São Paulo

```python
relacao_df = pd.DataFrame()
```

Com a funçao insert está pegando a coluna "Ano" do dataframe 1 e adicionado no novo dataframe

A função loc está criando uma coluna e completando com as informações da lista média

O ultimo insert está pegando a coluna "Produção_de_CO2" do primeiro dataframe e
copiando ela na terceira coluna do terceiro dataframe

```python
relacao_df.insert(0, 'Ano', df1['Ano'])
relacao_df.loc[:,"Media_Area_Plantada"] = media
relacao_df.insert(2, 'Produção_de_CO2', df1['Produção_de_CO2'])
```

```python
display(relacao_df)
```

Esta célula está produzindo um gráfico da média da área plantada de SP dos anos de 2013 a 2022

Ela utiliza funções do matplotlib, uma biblioteca específica para gráficos, a função "_plot_series" utiliza os valores do dataframe "relacao_df" pegando os dados da coluna "Ano" e armazenando no eixo X e os dados da coluna "Media_Area_Plantada" no eixo Y, tambem é utilizado da biblioteca seanborn para determinar a cor da linha do grafico e o comando plt.plot para exibir o grafico

As linhas fora da função determinam o estilo de gráfico e a forma que ele é apresentado

```python
def _plot_series(series, series_name, series_index=0):
  palette = list(sns.palettes.mpl_palette('Dark2'))
  xs = series['Ano']
  ys = series['Media_Area_Plantada']

  plt.plot(xs, ys, label=series_name, color=palette[series_index % len(palette)])

fig, ax = plt.subplots(figsize=(10, 5.2), layout='constrained')
df_sorted = relacao_df.sort_values('Ano', ascending=True)
_plot_series(df_sorted, '')
sns.despine(fig=fig, ax=ax)
plt.xlabel('Ano')
_ = plt.ylabel('Media_Area_Plantada')
```

Está célula faz a mesma coisa que a célula anterior, mas produz um gráfico da Produção de CO2 no estado de São Paulo nos anos de 2013 a 2022


```python
def _plot_series(series, series_name, series_index=0):
  palette = list(sns.color_palette("husl", 8))
  xs = series['Ano']
  ys = series['Produção_de_CO2']

  plt.plot(xs, ys, label=series_name, color=palette[series_index % len(palette)])

fig, ax = plt.subplots(figsize=(10, 5.2), layout='constrained')
df_sorted = relacao_df.sort_values('Ano', ascending=True)
_plot_series(df_sorted, '')
sns.despine(fig=fig, ax=ax)
plt.xlabel('Ano')
_ = plt.ylabel('Produção_de_CO2')
```

```python
def _plot_series(series, series_name, series_index=0):
    xs = series['Ano']
    ys = series['Media_Area_Plantada']
    ys2 = series['Produção_de_CO2']

    palette = list(sns.palettes.mpl_palette('Dark2'))

    fig, ax1 = plt.subplots(figsize=(10, 5.2), layout='constrained')

    ax1.plot(xs, ys, label='Média da Área Plantada', color=palette[series_index % len(palette)])
    ax1.set_xlabel('Anos')
    ax1.set_ylabel('Média da Área Plantada', color=palette[series_index % len(palette)])
    ax1.tick_params(axis='y', labelcolor=palette[series_index % len(palette)])

    ax2 = ax1.twinx()
    palette2 = list(sns.color_palette("husl", 8))
    ax2.plot(xs, ys2, label='Produção de CO2', color=palette2[series_index % len(palette2)])
    ax2.set_ylabel('Produção de CO2', color=palette2[series_index % len(palette2)])
    ax2.tick_params(axis='y', labelcolor=palette2[series_index % len(palette2)])

    fig.suptitle(series_name)

    ax1.legend(loc='upper left')
    ax2.legend(loc='upper right')

    plt.show()

df_sorted = relacao_df.sort_values('Ano', ascending=True)
_plot_series(df_sorted, 'Comparação entre Média da Área Plantada e Produção de CO2')
sns.despine()
```
